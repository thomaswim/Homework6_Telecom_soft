```python
# Example 2

import requests
url="https://www.alibaba.com/trade/search?fsb=y&IndexArea=product_en&CatId=&tab=all&SearchText=socket"

try:
    kv={'user-agent':'Mozilla/5.0'}
    r=requests.get(url,headers = kv)
    r.raise_for_status()
    r.encoding = r.apparent_encoding
    print(r.text[:2000])
except:
    print('anomali')
    
print(r.status_code)
print(r.encoding)
```

```python
# Example 2

import requests

keyword='Christmas socket'


try:
    kv={'q':keyword}
    r=requests.get("http://www.google.com/search",params = kv)
    print(r.request.url)
    r.raise_for_status()
    print(r.text)
except:
    print('anomali')
    
print(r.status_code)
print(r.encoding)
```

```python
import requests
from bs4 import BeautifulSoup

def getdata(url):
    r = requests.get(url)
    return r.text

htmldata = getdata('https://www.amazon.com/christmas-socks/s?k=christmas+socks')
soup = BeautifulSoup(htmldata, 'html.parser')
for item in soup.find_all('img'):
    print(item['src'])
```
